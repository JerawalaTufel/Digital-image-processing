```python
import cv2
import numpy as np

image = cv2.imread('socity.png')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow('Original image',image)
cv2.imshow('Gray image', gray)

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(image,-1,kernel)

cv2.imshow('Image after performing box filtering',dst)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from matplotlib import pyplot as plt

b1, g1, r1    = cv2.split(image)
b3, g3, r3    = cv2.split(dst)
#b2, g2, r2   = cv2.split(gray)

image1 = np.dstack((r1,g1,b1))
dst1 = np.dstack((r3,g3,b3))
#gray1=np.dstack((r2,g2,b2))

plt.subplot(121),plt.imshow(image1),plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst1),plt.title('Image after performing box filtering')
plt.xticks([]), plt.yticks([])

#plt.subplot(123),plt.imshow(gray1),plt.title('Image after performing gray scale')
#plt.xticks([]), plt.yticks([])

plt.show()
```
